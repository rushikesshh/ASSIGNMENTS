```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv("C:\\Users\\RUSHIKESH\\Downloads\\Fraud_check.csv")
data
```

```python
data.info()
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
data["Undergrad"] = label_encoder.fit_transform(data["Undergrad"])

data["Marital.Status"] = label_encoder.fit_transform(data["Marital.Status"])

data["Urban"] = label_encoder.fit_transform(data["Urban"])
```

```python
data
```

```python
data["taxable_income"] = data["Taxable.Income"].map(lambda a:'risky' if a<=30000 else 'good')
```

```python
data
```

```python
#data["taxable_income"] = label_encoder.fit_transform(data["taxable_income"])
```

```python
data1 = data.drop(["Taxable.Income"],axis=1)
```

```python
data1
```

```python
data1.info()
```

```python
# dividing the inputs and output :-

X = data1.iloc[:,:-1]
Y = data1.iloc[:,-1]
```

```python
# spliting the data into train and test :-

x_train, x_test , y_train, y_test = train_test_split(X,Y, test_size=0.2)
```

```python
y_train.shape
```

```python
y_test.shape
```

### Building the Decision Tree using Entropy crietria :-

```python
model = DecisionTreeClassifier(criterion='entropy')
model.fit(x_train, y_train)
```

```python
# plotting decision tree :-

#fn = ['Undergrad', 'Marital.Status', 'City.Population', 'Work.Experience', 'Urban']
#cn = ['good', 'risky']

#fig,axes = plt.subplots (nrows=1 , ncols=1, figsize=(4,3), dpi(800))

tree.plot_tree = (model);
```
